```python
relevance_train = relevance_train.dropna()
```

```python
relevance_train['Relevance'] = [int(i) for i in relevance_train['Relevance']]
relevance_train['DocumentId'] = [int(i) for i in relevance_train['DocumentId']]
```

```python
doc_ids = np.unique(relevance_train['DocumentId'])
```

```python
doc_ids
```

```python
for i in doc_ids:
    relevance_train.loc[relevance_train['DocumentId']==i, 'AvgRel'] = np.sum(relevance_train.loc[relevance_train['DocumentId']==i, 'Relevance'])/len(relevance_train.loc[relevance_train['DocumentId']==i, 'Relevance'])
```

```python
relevance_train.head()
```

```python
relevance_train.sort_values('AvgRel', ascending=False)
```

```python
rel_df = pd.DataFrame(np.transpose([doc_ids, np.zeros(np.shape(doc_ids))]), columns=['DocumentId', 'AvgRel'])
```

```python
rel_df.head()
```

```python
for i in rel_df['DocumentId']:
    rel_df.loc[rel_df['DocumentId']==i, 'AvgRel'] = np.max(relevance_train.loc[relevance_train['DocumentId']==i, 'AvgRel'])
```

```python
rel_df = rel_df.sort_values('AvgRel', ascending=False)
```
