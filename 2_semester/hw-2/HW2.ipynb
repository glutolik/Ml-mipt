```python
data = ts_dep42['item 42356'].to_frame()
```

```python
data.plot(figsize=(15,7))
```

```python
data = data.dropna()
data.plot(figsize=(15,7))
```

```python
data['stationary'] = stats.boxcox(data['item 42356'] + 0.01)[0]
```

```python
data['stationary'].plot(figsize=(15,7))
```

```python
kpss_stat = stattools.kpss(np.array(data['stationary']))
print('The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss_stat[0], kpss_stat[1], kpss_stat[2], kpss_stat[3]))
```

```python
df_stat = stattools.adfuller(np.array(data['stationary']))
print('The Augmented Dickey-Fuller test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ncritical values: %s' % 
      (df_stat[0], df_stat[1], df_stat[2], df_stat[4]))
```

#### TS is stationary

```python
train = data['stationary'].head(round(0.8 * len(data)))
test = data['stationary'].tail(round(0.2 * len(data))).to_frame()
```

```python
fig = plt.figure(figsize=(15,7))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(train, lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(train, lags=40, ax=ax2)
fig.tight_layout()
```

```python
best_p = 0
best_q = 0
best_aic = 4000
for i in range(4):
    for j in range(4):
        arma_model = sm.tsa.ARMA(train, (i, j)).fit()
        print('(', i, ',', j, ')')
        #print(arma_mod00.params)
        print(arma_model.aic)
        if(arma_model.aic < best_aic):
            best_aic = arma_model.aic
            best_p = i
            best_q = j
```

```python
print(best_aic, best_p, best_q)
```

```python
model = sm.tsa.ARMA(train, (best_p, best_q)).fit()
```

```python
model.resid.plot(figsize=(15,4));
```

```python
kpss_stat= stattools.kpss(model.resid.values)
print('The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss_stat[0], kpss_stat[1], kpss_stat[2], kpss_stat[3]))
```

```python
df_stat = stattools.adfuller(model.resid.values)
print('The Augmented Dickey-Fuller test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ncritical values: %s' % 
      (df_stat[0], df_stat[1], df_stat[2], df_stat[4]))
```

```python
pd.DataFrame({'lags':range(30), 'pvalue':sm.stats.diagnostic.acorr_ljungbox(model.resid.values, lags=30)[1], 
              'critial':np.array([0.05]*30)}).set_index('lags').plot(figsize=(15,5))
```

```python
fig = plt.figure(figsize=(15,7))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(model.resid, lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(model.resid, lags=40, ax=ax2)
fig.tight_layout()
```

```python
def loss_MACAPE(x,y):
    MACAPE_sum = 2 * np.abs((x - y)) / (x + y)
    return np.mean(MACAPE_sum)
```

```python
def loss_SMAPE(x, y):
    SMAPE_sum = np.abs(x - y) / (np.abs(x) + np.abs(y))
    return np.mean(SMAPE_sum)
```

```python
print(loss_SMAPE(test['forecast'],test['stationary']))
```

**II. Find the best parameters of ARIMA automaticaly** 
* 0) read data for TS under consideration and use the first 80% of all length of the TS as train TS.  
* 1) plot the graph of the train TS, preprocess data (delete tail, make calendar transformation if needed)
* 2) stabilize dispersion of train TS through box-cox transformation if needed.
* 3) Run grid search for ARIMA parameters (use transformed train TS) according to AIC. Use at least 10 distinct values for $p,q,d$. How many time does it take?
* 4) Build forecast for the initial TS with best ARIMA model selected automaticaly at step 2 (you need to transform it like in step 2).
* 5) Build forecast interval for the last 20% of transformed TS. Calculate loss MACAPE of your forecast. Does the forecast look appropriate?

```python
best_p_2 = 0
best_d_2 = 0
best_q_2 = 0
best_aic = 4000
for p in range(10):
    for d in range(1):
        for q in range(10):
            try:
                arima_model = sm.tsa.ARIMA(train, (p, d, q)).fit()
                print('(', p, ',', d, ',', q, ')')
                #print(arma_mod00.params)
                print(arima_model.aic)
                if(arima_model.aic < best_aic):
                    best_aic = arima_model.aic
                    best_p_2 = p
                    best_d_2 = d
                    best_q_2 = q
            except:
                continue
```

```python
print(best_aic, best_p_2, best_d_2, best_q_2)
```

```python
new_model = sm.tsa.ARIMA(train, (best_p_2, best_d_2, best_q_2)).fit()
```

```python
new_model.resid.plot(figsize=(15,4));
```

```python
new_pred = new_model.predict()
```

```python
plt.figure(figsize=(15,7))
(train - train.shift()).plot()
new_pred.plot()
```

```python
new_forecast = new_model.forecast(steps=len(test))[0]
print(new_forecast)
```
