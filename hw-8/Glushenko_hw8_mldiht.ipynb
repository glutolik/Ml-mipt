```python
import numpy as np
import pylab as plt
from numpy.linalg import svd
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * np.dot(F, np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = svd(Sigma, full_matrices=False)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    sorted_id = d.argsort()[::-1]
    sorted_val = d[sorted_id]
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / \
                dTot * 100.0 for i in range(n)])
    part_sum = 0
    for i in range(len(sorted_val)):
        part_sum += sorted_val[i]
        if(part_sum/dTot >= varRetained):
            m = i + 1
            break
    print '%.2f %% variance retained in %m dimensions', m, part_sum/dTot

    # plot the variance plot
    if show:
        plt.figure(figsize = (15, 5))
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = (U.T[sorted_id[:m]]).T
    G = np.dot(np.transpose(F), U_reduced)

    return G, U_reduced
```

```python
import numpy as np

from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_mutual_info_score, adjusted_rand_score
```

```python
pca = PCA(n_components=2)

pca.fit(X)
X_test_pca = pca.transform(X)

db = DBSCAN(eps=22, min_samples=10).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
print("Adjusted Rand Index: %0.3f"
      % metrics.adjusted_rand_score(y, labels))
print("Adjusted Mutual Information: %0.3f"
      % metrics.adjusted_mutual_info_score(y, labels))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Black removed and is used for noise instead.
plt.figure(figsize = (15, 7))
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = X_test_pca[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)

    xy = X_test_pca[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=6)

    
plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```
